```python
import pandas as pd
df = pd.read_csv('aaaaa_1.csv')
```

```python
# 使用布尔索引筛选出PY列中大于或等于2008且小于2023的行
df = df[(df['aa'] >= 2008) & (df['aa'] < 2024)]
#删除空值所在的行
df.dropna(inplace=True)
```

```python
categorys = df['bb']
categorys
```

```python
year = df['aa']
categoryPerYear = [categorys,year]
result = pd.concat(categoryPerYear, axis=1)
result.to_csv('task3_categoryPerYear.csv')
top10Category = categorys.value_counts()[:10]
top10Category.to_csv('task3_top10Category.csv')
```
