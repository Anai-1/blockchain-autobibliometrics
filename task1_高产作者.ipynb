```python
import pandas as pd
df = pd.read_csv('task3.csv')
# 使用布尔索引筛选出PY列中大于或等于2008且小于2023的行
df = df[(df['PY'] >= 2008) & (df['PY'] < 2024)]
#删除空值所在的行
df.dropna(inplace=True)
```

```python
# 提取AF列并创建新的AF1列
df['AF1'] = df['AF'].apply(lambda x: x.split('|')[0] if '|' in x else x)

# 计算AF1列中每个值的出现次数
af1_counts = df['AF1'].value_counts()

# 将计数结果添加到DataFrame中
df['AF1_Count'] = df['AF1'].map(af1_counts)
```

```python
# 保存结果到新的CSV文件
df.to_csv('task2_topscientist.csv', index=False)
df
```

```python
import pandas as pd

# 读取修改后的CSV文件（包含AF1和AF1_Count列）
df = pd.read_csv('task1_topscientist.csv')

# 使用value_counts方法获取AF1列中每个值的频次并选择前15个
top_15_af1_counts = df['AF1'].value_counts().head(20)

# 打印排名前15的AF1及其频次
print("排名前15的AF1与它们的频次:")
for af1, count in top_15_af1_counts.items():
    print(f"AF1: {af1}, 频次: {count}")
```
