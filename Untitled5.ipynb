```python
import pandas as pd
import matplotlib.pyplot as plt
import metaknowledge as mk

RC = mk.RecordCollection('data')
selectedTags = ['PY', 'Z9']
RC.writeCSV('RPYS_1.csv',onlyTheseTags=selectedTags)
```

```python
# 读取 WoS 数据文件
df = pd.read_csv('data.csv')
```

```python
df = df[['PY', 'Z9']]
```

```python
wos_df = df
```

```python
# 对文献按照出版年份分组，并计算每个分组的被引用次数平均值
wos_df = wos_df.groupby(by='PY', as_index=False).mean()
wos_df['PY'] = pd.to_datetime(wos_df['PY'], format='%Y')

# 计算前 n 年的被引用次数平均值
n_years = 10  # 指定前 10 年的被引用次数平均值
for i in range(n_years, len(wos_df)):
    start_year = wos_df.iloc[i-n_years]['PY'].year
    end_year = wos_df.iloc[i-1]['PY'].year
    mean_cites = wos_df[(wos_df['PY'].dt.year >= start_year) & 
                        (wos_df['PY'].dt.year < end_year)]['Z9'].mean()
    wos_df.loc[i, 'Mean Cites'] = mean_cites
```

```python
import numpy as np
# 计算角点
diffs = wos_df.iloc[n_years:]['Z9'] - wos_df.iloc[n_years:]['Mean Cites']
idx_min, idx_max = np.argmin(diffs.values), np.argmax(diffs.values)

# 绘制角点分析图
import matplotlib.pyplot as plt
plt.plot(wos_df.iloc[n_years:]['PY'], diffs)
plt.plot(wos_df.iloc[n_years+idx_min]['PY'], diffs.iloc[idx_min], 'ro')
plt.plot(wos_df.iloc[n_years+idx_max]['PY'], diffs.iloc[idx_max], 'go')
plt.show()
```
