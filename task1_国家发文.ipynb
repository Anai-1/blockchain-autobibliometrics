```python
import pandas as pd
```

```python
pf = pd.read_csv("output1.csv")
pf
```

```python
# 使用布尔索引筛选出PY列中大于或等于2008且小于2023的行
filtered_pf = pf[(pf['PY'] >= 2008) & (pf['PY'] < 2023)]
filtered_pf
```

```python
#删除空值所在的行
filtered_pf.dropna(inplace=True)
```

```python
# 提取RP列并重命名为RP'
filtered_pf['RP\''] = filtered_pf['RP'].str.split().str[-1]
```

```python
# 提取其余PY列
py_columns = [col for col in filtered_pf.columns if col.startswith('PY')]
```

```python
# 创建新的DataFrame，包含RP'和PY列
new_df = filtered_pf[['RP\''] + py_columns]

# 保存结果到新的CSV文件
new_df.to_csv('国家&年份.csv', index=False)
```
