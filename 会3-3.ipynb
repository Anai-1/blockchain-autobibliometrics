```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('会_国家&年份.csv')
countrys = df['RP\'']
year = df['PY']
```

```python
i = 0
for country in countrys:
    RP_words = str(countrys[i]).split()
    country = RP_words[-1]
    country = country.rstrip('.')
    countrys[i] = country
    i += 1
```

```python
countrys
```

```python
df['RP'] = countrys
countryPerYear = [countrys,year]
result = pd.concat(countryPerYear, axis=1)
result.to_csv('会_countryPerYear.csv')
top10Country = countrys.value_counts()[:10]
top10Country.to_csv('会_top10Country.csv')
```
