```python
import pandas as pd
```

```python
journals = [
    
]

# 现在，你有一个包含所有期刊名称的列表"journals"
df = pd.read_csv("task3.csv")
# 将J9列和journals列表中的值都转换为小写
df['J9'] = df['J9'].str.lower()
journals = [journal.lower() for journal in journals]
# 使用不在journals列表中的值筛选DataFrame
filtered_df = df[~df['J9'].isin(journals)]
# 使用布尔索引筛选出PY列中大于或等于2008且小于2023的行
filtered_df = filtered_df[(filtered_df['PY'] >= 2008) & (filtered_df['PY'] < 2024)]
#删除空值所在的行
filtered_df.dropna(inplace=True)
filtered_df
```

```python
# 提取RP列并重命名为RP'
filtered_df['RP\''] = filtered_df['RP'].str.split().str[-1]

# 提取其余PY列
py_columns = [col for col in filtered_df.columns if col.startswith('PY')]

# 创建新的DataFrame，包含RP'和PY列
new_df = filtered_df[['RP\''] + py_columns]

# 保存结果到新的CSV文件
new_df.to_csv('会_国家&年份.csv', index=False)

df1 = pd.read_csv('会_国家&年份.csv')
df1
```
