```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('task2_国家&年份.csv')
countrys = df['RP\'']
year = df['PY']
```

```python
i = 0
for country in countrys:
    RP_words = str(countrys[i]).split()
    country = RP_words[-1]
    country = country.rstrip('.')
    countrys[i] = country
    i += 1
```

```python
countrys
```

```python
df['RP'] = countrys
countryPerYear = [countrys,year]
result = pd.concat(countryPerYear, axis=1)
result.to_csv('task2_countryPerYear.csv')
top10Country = countrys.value_counts()[:10]
top10Country.to_csv('task2_top10Country.csv')
```

```python
df_1 = pd.read_csv('task2_countryPerYear.csv')
test = df_1.pivot_table(index=['RP\'','PY'],values=['sum'],aggfunc={'sum':np.sum},fill_value=0)
test.to_excel('task2_test.xlsx')
readtest = pd.read_excel('task2_test.xlsx')
RP = readtest['RP\'']
readtop10country = pd.read_csv('task2_top10Country.csv',index_col=0)
#top10 = readtop10country['Unnamed: 0']
print(readtop10country)
```
