```python
import pandas as pd
df = pd.read_csv("title_year_author_instation.csv")
```

```python
# 将J9列和journals列表中的值都转换为小写
df['J9'] = df['J9'].str.lower()
journals = [journal.lower() for journal in journals]
# 使用不在journals列表中的值筛选DataFrame
filtered_df = df[~df['J9'].isin(journals)]
filtered_df
```

```python
# 使用布尔索引筛选出PY列中大于或等于2008且小于2024的行
filtered_df = filtered_df[(filtered_df['PY'] >= 2008) & (filtered_df['PY'] < 2024)]
filtered_df
```

```python
#删除空值所在的行
filtered_df.dropna(inplace=True)
filtered_df
```

```python
import numpy as np
year = filtered_df['PY']
data1 = np.array(year)
```

```python
counts = pd.value_counts(data1).sort_index(ascending=True)
print(counts)
```
