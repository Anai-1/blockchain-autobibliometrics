```python
import pandas as pd
```

```python
pf = pd.read_csv("title_year_author_instation.csv")
pf
```

```python
# 使用布尔索引筛选出PY列中大于或等于2008且小于2023的行
filtered_pf = pf[(pf['PY'] >= 2008) & (pf['PY'] < 2023)]
```

```python
filtered_pf
```

```python
import numpy as np
year = filtered_pf['PY']
data1 = np.array(year)
```

```python
counts = pd.value_counts(data1).sort_index(ascending=True)
print(counts)
```
